**PSNR SCORE BETWEEN GENERATED AND REFERENCE-3k**

```python
from skimage.metrics import peak_signal_noise_ratio as psnr
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt

def calculate_psnr(reference_path, generated_path, display_image=False):
    # Read images
    ref_image = cv2.imread(reference_path, cv2.IMREAD_COLOR)
    gen_image = cv2.imread(generated_path, cv2.IMREAD_COLOR)

    if ref_image is None or gen_image is None:
        print(f"Error reading images: {reference_path}, {generated_path}")
        return None

    # Resize generated image to match reference image dimensions
    gen_image_resized = cv2.resize(gen_image, (ref_image.shape[1], ref_image.shape[0]))

    # Display the resized generated image if requested
    if display_image:
        plt.imshow(cv2.cvtColor(gen_image_resized, cv2.COLOR_BGR2RGB))
        plt.title("Resized Generated Image")
        plt.axis("off")
        plt.show()

    # Compute PSNR
    return psnr(ref_image, gen_image_resized)

# Normalize filenames (remove extensions and convert to lowercase)
def normalize_filename(filename):
    return os.path.splitext(filename)[0].lower()

# Calculate average PSNR for paired images
def process_psnr(real_folder, generated_folder):
    real_images = {normalize_filename(f): os.path.join(real_folder, f)
                   for f in os.listdir(real_folder) if f.endswith(('.jpg', '.png', '.jpeg'))}
    generated_images = {normalize_filename(f): os.path.join(generated_folder, f)
                        for f in os.listdir(generated_folder) if f.endswith(('.jpg', '.png', '.jpeg'))}

    # Find matches
    matching_files = set(real_images.keys()) & set(generated_images.keys())
    print(f"Found {len(matching_files)} matching files.")

    if not matching_files:
        print("No matching files found.")
        return float('nan')

    # Compute PSNR for each matching pair
    psnr_scores = []
    for i, filename in enumerate(matching_files):
        ref_path = real_images[filename]
        gen_path = generated_images[filename]

        # Display the first resized generated image
        psnr_value = calculate_psnr(ref_path, gen_path, display_image=(i == 0))
        if psnr_value is not None:
            psnr_scores.append(psnr_value)

    if not psnr_scores:  # Handle the case of no valid comparisons
        print("No valid PSNR scores computed.")
        return float('nan')

    return np.mean(psnr_scores)

# Example usage
real_folder = "/content/images/images_200"
generated_folder = "/content/output/best_generated_images"

psnr_score = process_psnr(real_folder, generated_folder)
if not np.isnan(psnr_score):
    print(f"PSNR Baseline: {psnr_score}")
else:
    print("PSNR calculation failed: No valid comparisons or matches.")
```

```python
import os
import torch
from PIL import Image
from torchmetrics.functional.multimodal import clip_score
from functools import partial
import re
import numpy as np

# Initialize the CLIP scoring function
clip_score_fn = partial(clip_score, model_name_or_path="openai/clip-vit-base-patch16")

def preprocess_image(image_path):
    try:
        # Load and normalize the image (rescaling to [0, 1])
        image = Image.open(image_path).convert("RGB")
        image = image.resize((224, 224))  # Resize to the size required by CLIP
        image = torch.tensor(np.array(image) / 255.0).permute(2, 0, 1)  # Convert to tensor and rearrange dimensions
        return image.unsqueeze(0)  # Add batch dimension
    except Exception as e:
        print(f"Error loading image {image_path}: {e}")
        return None

def extract_digits(filename):
    match = re.search(r'\d+', filename)
    return match.group() if match else None

def calculate_clip_score_with_text_files(image_folder, text_folder):
    # Extract digits from filenames in the image folder
    image_files = os.listdir(image_folder)
    image_digits = {extract_digits(f): f for f in image_files if extract_digits(f)}

    # Extract digits from text files in the text folder
    text_files = os.listdir(text_folder)
    text_digits = {extract_digits(f): f for f in text_files if extract_digits(f)}

    # Find matches based on the numeric part of filenames
    matching_digits = set(image_digits.keys()) & set(text_digits.keys())
    print(f"Found {len(matching_digits)} matching image-text pairs.")

    if len(matching_digits) == 0:
        print("No matching image-text pairs found.")
        return float('nan')

    # Prepare lists to hold images and text descriptions
    images = []
    prompts = []

    for digits in matching_digits:
        image_path = os.path.join(image_folder, image_digits[digits])
        text_path = os.path.join(text_folder, text_digits[digits])

        # Read text prompt
        try:
            with open(text_path, 'r') as f:
                description = f.read().strip()
        except Exception as e:
            print(f"Error reading text file {text_path}: {e}")
            continue

        # Preprocess image
        preprocessed_image = preprocess_image(image_path)
        if preprocessed_image is not None:
            images.append(preprocessed_image)
            prompts.append(description)

    if not images:
        print("No valid image-text pairs to process.")
        return float('nan')

    # Stack all images into a tensor
    images_tensor = torch.cat(images, dim=0)

    # Compute CLIP scores
    clip_scores = clip_score_fn(images_tensor, prompts).detach()

    # Calculate and return the average score
    average_clip_score = torch.mean(clip_scores).item()
    return round(average_clip_score, 4)

# Example usage
image_folder = "/content/o_2k"  # Path to folder with images
text_folder = "/content/processed_texts/processed_texts"  # Path to folder with text files

average_clip_score = calculate_clip_score_with_text_files(image_folder, text_folder)
print(f"Average CLIP score: {average_clip_score}")
```
