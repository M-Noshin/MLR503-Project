```python
import PIL
import requests
import torch
from diffusers import StableDiffusionInstructPix2PixPipeline, EulerAncestralDiscreteScheduler

model_id = "/content/drive/MyDrive/tuned_model_final_2000steps"
pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(model_id, torch_dtype=torch.float16, safety_checker=None)
pipe.to("cuda")
pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
```

```python
import os
from PIL import Image

# Define folder paths
sketch_folder = '/content/test_sketches/test_sketches'
text_folder = '/content/processed_texts/processed_texts'
output_folder = './output_faces'

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Get all sketch filenames
sketch_files = [f for f in os.listdir(sketch_folder) if f.endswith('.jpg')]

# Process each sketch
for sketch_file in sketch_files:
    # Get the file name without extension to match with text files
    file_name = os.path.splitext(sketch_file)[0]

    # Construct full paths for sketch, text, and output
    sketch_path = os.path.join(sketch_folder, sketch_file)
    text_path = os.path.join(text_folder, f'{file_name}.txt')
    output_path = os.path.join(output_folder, f'{file_name}.png')

    # Check if the corresponding text file exists
    if not os.path.exists(text_path):
        print(f"Text file missing for sketch: {sketch_file}. Skipping...")
        continue

    # Load sketch and text prompt
    input_image = Image.open(sketch_path).convert("RGB")
    with open(text_path, 'r') as file:
        prompt = file.read().strip()

    # Disable the safety checker for inference
    pipe.safety_checker = None

    # Generate the face image
    try:
        generated_images = pipe(prompt=prompt, image=input_image, strength=0.4, guidance_scale=7).images
        generated_images[0].save(output_path)
        print(f"Generated image saved to: {output_path}")
    except Exception as e:
        print(f"Error generating image for sketch {sketch_file}: {e}")

print("All images have been processed and saved.")
```

```python
import shutil

# Define the folder to compress and the output .zip file name
folder_to_compress = '/content/output_faces'
zip_file = 'output_faces_2000steps.zip'

# Compress the folder into a zip file
shutil.make_archive(base_name='output_faces', format='zip', root_dir=folder_to_compress)

print(f"Folder '{folder_to_compress}' has been compressed into '{zip_file}'.")
```
